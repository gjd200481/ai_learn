```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset = pd.read_csv('../data/studentscores.csv')
X = dataset.iloc[ : , :1].values
Y = dataset.iloc[ : ,1].values

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, 
                                                    test_size=1/4, 
                                                    random_state=0)
```

简单的线性回归模型

dataset.iloc[:, :1] = 取所有行的、从第0列开始到第1列之前结束的列 → 即只取第0列，但保持二维形状 (行数, 1)，作为特征矩阵 X。
dataset.iloc[:, 1] = 取所有行的第2列（索引为1的列），结果为一维数组 (n,)，作为目标值 Y。

from sklearn.model_selection import train_test_split 
把数据 X 和标签 y 分成训练集和测试集
其中 1/4 用来测试，80% 用来3/4
random_state 固定随机数种子，让每次随机划分结果都一样。

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor = regressor.fit(X_train, Y_train)
```

sklearn.linear_model包实现了广义线性模型，包括线性回归、Ridge回归、Bayesian（贝叶斯）回归等。
LinearRegression是其中较为简单的线性回归模型。


```python
Y_pred = regressor.predict(X_test)
```

```python
plt.scatter(X_train, Y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')

plt.show()
```

```python
plt.scatter(X_test, Y_test, color='red')
plt.plot(X_test, regressor.predict(X_test), color='blue')
plt.show()
```
